# Graph of Occupation vs Average age groups

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Occupation',
    y='Age',
    data=df,
    estimator='mean',
    palette='viridis'
)

plt.title("Average Age by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Average Age")
plt.xticks(rotation=45)
plt.show()
```

# Graph of averge stress level by age group

```python
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Age Group',
    y='Stress Level',
    data=df,
    estimator='mean',
     palette='coolwarm'
)

plt.title("Average Stress Level by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Stress Level")
plt.show()
```

# Graph of Average Sleep Duration by Age Group

```python
df['Age Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 55, 65],
    labels=['18-25', '26-35', '36-45', '46-55', '56-65']
)
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Age Group',
    y='Sleep Duration',
    data=df,
    estimator='mean',
     palette='Purples'
    
)

plt.title("Average Sleep Duration by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Sleep Duration (hours)")
plt.show()
```

# Graph of average quality of sleep by age group

```python
df['Age Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 55, 65],
    labels=['18-25', '26-35', '36-45', '46-55', '56-65']
)
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Age Group',
    y='Quality of Sleep',
    data=df,
    estimator='mean',
     palette='magma'
)

plt.title("Average Quality of Sleep by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Quality of Sleep")
plt.show()
```

# Graph of Average Sleep Duration by Sleep Disorder

```python
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Sleep Disorder',
    y='Age',
    data=df,
    estimator='mean',
     palette='Reds'
)

plt.title("Average Sleep Duration by Sleep Disorder")
plt.xlabel("Sleep Disorder")
plt.ylabel("Average Sleep Duration (hours)")
plt.show()

```

# Proportion of average physical activity 

```python
df['Age Group'] = pd.cut(
    df['Age'],
    bins=[0, 25, 35, 45, 55, 100],   # wider range
    labels=['≤25', '26-35', '36-45', '46-55', '56+']
)
age_activity = (
    df.groupby('Age Group')['Physical Activity Level']
    .mean()
    .dropna()
)

age_activity
plt.figure(figsize=(7, 7))
plt.pie(
    age_activity.values,
    labels=age_activity.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Proportion of Average Physical Activity Level by Age Group")
plt.show()

```

# Comparision of sleep disorder based on gender

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Gender', hue='Sleep Disorder')
plt.title('Gender vs Sleep Disorder')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Sleep Disorder')
plt.show()
```
